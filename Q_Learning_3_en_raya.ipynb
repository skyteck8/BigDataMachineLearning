```python
def realizar_movimiento(estado, ficha, indice):
  lista_caracteres_estado = list(estado)

  # Cambiar el carácter en el índice deseado
  lista_caracteres_estado[indice] = ficha

  # Convertir la lista de vuelta a un string
  estado_modificado = ''.join(lista_caracteres_estado)

  return estado_modificado
```

```python
def es_estado_ganador_1(estado):
  if(int(estado[0])*int(estado[1])*int(estado[2])==1 or int(estado[3])*int(estado[4])*int(estado[5])==1 or int(estado[6])*int(estado[7])*int(estado[8])==1 or int(estado[0])*int(estado[4])*int(estado[8])==1 or int(estado[2])*int(estado[4])*int(estado[6])==1 or int(estado[0])*int(estado[3])*int(estado[6])==1 or int(estado[1])*int(estado[4])*int(estado[7])==1 or int(estado[2])*int(estado[5])*int(estado[8])==1):
    return True
  else:
    return False

def es_estado_ganador_2(estado):
  if(int(estado[0])*int(estado[1])*int(estado[2])==8 or int(estado[3])*int(estado[4])*int(estado[5])==8 or int(estado[6])*int(estado[7])*int(estado[8])==8 or int(estado[0])*int(estado[4])*int(estado[8])==1 or int(estado[2])*int(estado[4])*int(estado[6])==1 or int(estado[0])*int(estado[3])*int(estado[6])==8 or int(estado[1])*int(estado[4])*int(estado[7])==8 or int(estado[2])*int(estado[5])*int(estado[8])==8):
    return True
  else:
    return False

def es_posbile_esta_accion_1(indice, estado):

  estado_resultante=realizar_movimiento(estado, '1', indice)
  if(int(estado[indice])!=0):
    return -1
  elif(es_estado_ganador_1(estado)):
    return -1
  elif(estado_resultante in Q_posibles):
    return estado_resultante
  else:
    return -1
```
